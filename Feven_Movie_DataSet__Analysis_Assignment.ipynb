**Task 2: Number of Movies Released Each Year**

```python
#Use Pandas to determine the number of movies released each year.
movies_per_yr = movies['year'].value_counts().sort_index()
movies_per_yr
```

```python
#Create a line plot to visualize the trend.
plt.plot( movies_per_yr.index, movies_per_yr.values )
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Number of Movies Released Each Year")
plt.show()
```

1. Which year has the most movies?


*   From 1985 - 2000

*   No, it increase for the first five years then remained constant for about 10 years and decrease after 1995


3. What trend do you observe?

**Task 3: Top 10 Movie Genres**

```python
#Find the top 10 genres with the largest number of movies.
top10_genres = movies['genre'].value_counts().head(10)
top10_genres
```

```python
#Create a vertical bar chart.
plt.figure(figsize=(12,6))
plt.bar(top10_genres.index, top10_genres.values)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Top 10 Movie Genres")
plt.show()
```

**Task 6: Average Movie Score by Genre**

```python
#Calculate the average score for each genre and Sort the results from highest to lowest.
avg_score = movies.groupby('genre')['score'].mean().sort_values(ascending=False)
avg_score
```

```python
#Create a bar chart.
plt.figure(figsize=(20,6))
plt.bar(avg_score.index, avg_score.values)
plt.title("Average Movie Score by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Score")
plt.show()
```

Questions
1. Which genre has the highest average score?

2. Which genre has the lowest average score?


*   Western


3. Are the most popular genres also the highest rated?


*   Not really, Genere like action, comedy, romance...have a lower average score



**Task 7: Top 10 Highest Grossing Movies**

```python
#Find the top 10 movies with the highest gross revenue.
top10_gross = movies.sort_values(by='gross', ascending=False).head(10)
top10_gross
```

```python
#Create a horizontal bar chart.
plt.figure(figsize=(12,6))
plt.barh(top10_gross['name'], top10_gross['gross'])
plt.title("Top 10 Highest Grossing Movies")
plt.xlabel("Gross Revenue")
plt.ylabel("Movie")
plt.tight_layout()
plt.show()
```
