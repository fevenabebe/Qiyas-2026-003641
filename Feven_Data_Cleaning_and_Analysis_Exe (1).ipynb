# Pandas Assignment on Data Cleaning and Analysis

```python
# import libraries
import pandas as pd
import numpy as np

# Prepare the dataframe
university=pd.DataFrame({
    "student_id":[201,202,203,204,205,206,207,208,209,210],
    "name":["Abel", "Sara", None, "Jhon", "Martha", None, "David", "Helen", "Tom", None],
    "department":["CT","IT","CS", None,"SE","IT", "SE", None,"CS","IT"],
    "gpa":[3.5, 3.8, np.nan, 2.9, 3.7, np.nan, 3.2, 3.4, 3.9, np.nan],
    "scholarship": [5000, np.nan, 3000, 2000, np.nan, 4000, 3500, np.nan, 5000, np.nan]
})
```

Data Cleaning

```python
# Count missing values
Missing_count= university.isnull().sum()
Missing_count
```

```python
# Calculate missing-value percentage
percentage = (Missing_count/len(university))*100
percentage
```

```python
# Replace missing names with unknown
university["name"]=university["name"].fillna("Unknown")
university["name"]
```

```python
# Replace missing department with mode
university["department"]=university["department"].fillna(university["department"].mode()[0])
university["department"]
```

```python
#Replace missing gpa values with median
university["gpa"]=university["gpa"].fillna(university["gpa"].median())
university["gpa"]
```

```python
#Replace missing scholarship with zero
university["scholarship"]=university["scholarship"].fillna(0)
university["scholarship"]
```

```python
#Data Analysis
```

```python
#Count students by department
counts = {}

for index, row in university.iterrows():

    dept = row["department"]

    counts[dept] = counts.get(dept, 0) + 1
print(counts)
```

```python
#count scholrship recipients
scholarship_recipients = university['scholarship'].count()
print(f"\nNumber of scholarship recipients: {scholarship_recipients}")
```

```python
#Find highest gpa
highest_gpa = university['gpa'].max()
print( f"The highest GPA is {highest_gpa}")
```

```python
# Find the lowest gpa
lowest_gpa = university['gpa'].min()
print( f"The lowest GPA is {lowest_gpa}")
```

```python
# Calculate average gpa
average_gpa = university['gpa'].mean()
print( f"The average GPA is {average_gpa}")
```

```python
# Calculate total scholarship amount
total_scholarship=university["scholarship"].sum()
print( f"The total Scholarship amonunt is {total_scholarship}")
```
