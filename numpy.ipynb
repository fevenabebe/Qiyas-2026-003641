```python
#library import
import numpy as np
```

```python
arr=[[1,2,3],[4,5,6]]
result= arr*2
print (result)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
result= arr*2
print (result)
```

```python
arr = np.array([[-1, 2, 0, 4, 6], 
                [4, -0.5, 6, 0,9], 
                [2.6, 0, 7, 8,4], 
                [3, -7, 4, 2.0,3]]) 
print("Dimension of array",arr.ndim)
print("Shape of array",arr.shape)
print()
print("Transpose of the array",arr.T, "\n")
print("Element on row 2 and column 2: ",arr[2][2], "\n")
print("Elements start from 0 to 4 , with three jump of row",arr[:4][::3], "\n")
print("Elements in range of row 0 and 4, with two jump of column",arr[0:4,::2], "\n")
```

```python
#Reshape arrray
reshaped = arr.reshape(2,10)
print(reshaped)
```

```python
#Copy array
arr1= reshaped.copy()
print(arr1)
```

```python
#3D array
a=np.array([[[1,2],[2.3,4], [4,5]]])
print("Dimention of array a :",a.ndim)
print("Data type of array a :",a.dtype)
print("Sum of a =", a.sum())
print("Square root of a =", np.sqrt(a),"\n")

b=np.array([[[6,7],[8,9],[10,11]]])
print("Dimention of array b :",b.ndim)
print("Data type of array b :",b.dtype)
print("Sum of b =", np.sum(b))
print("Square root of b =", np.sqrt(a), "\n")
print("Sum of a and b", a+b)

```
