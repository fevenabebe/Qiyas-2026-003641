```python
import pandas as pd
import numpy as np
students = pd.DataFrame({
"student_id":[101,102,103,104,105,106,107,108,109,110],
"name":
["Abel","Sara",None,"John","Marta",None,"David","Helen","Tom",None],
"department":["CS","IT","CS",None,"SE","IT","SE",None,"CS","IT"],
"gpa":[3.5,3.8,np.nan,2.9,3.7,np.nan,3.2,3.4,3.9,np.nan],
"scholarship":
[5000,np.nan,3000,2000,np.nan,4000,3500,np.nan,5000,np.nan]
})
students
```

```python
#Missing value
students.isnull()
```

```python
#count missing value
students.isnull().sum()
```

```python
#Total number of missing value
students.isnull().sum().sum()
```

```python
#Percentage of missing value - help to determine whether a column should be cleaned, imputed, or removed.
(students.isnull().sum())/len(students)*100
```

```python
#Display students whose GPA is missing.
students[students["gpa"].isnull()]
```

```python
#Display students whose scholarship information is missing.
students[students["scholarship"].isnull()]
```

```python
#Exercise 9
#Replace missing names with "Unknown".
students["name"]=students["name"].fillna("Unknown")
students["name"]
```

```python
#Exercise 10
#Replace missing GPA values using the mean GPA.
students["gpa"]=students["gpa"].fillna(students["gpa"].mean())
students["gpa"]
```

```python
#Exercise 11
#Replace missing GPA values using the median GPA.
students["gpa"]=students["gpa"].fillna(students["gpa"].median())
students["gpa"]
```

```python
#Exercise 12
#Replace missing department values using the most common department.
students["department"]=students["department"].fillna(students["department"].mode()[0])
students["department"]
```

```python
#Exercise 13
#Replace missing scholrship values with zero
students["scholarship"] = students["scholarship"].fillna(0)
students["scholarship"]
```

```python
#Exercise 14
#remove missing rows
students.dropna()
```

```python
#Exercise 15
#Remove missing columns
students.dropna(axis=1)
```

```python
#Exercise 16
#Create a GPA missing-value flag.
students["gpa_missing"] = students["gpa"].isnull()
students["gpa_missing"]
```

```python
#Exercise 17
#Create a scholarship missing-value flag.
students["scholarship_missing"] = students["scholarship"].isnull()
students["scholarship_missing"]
```

```python
#Exercise 18
#Find the column with the highest number of missing values.
students.isnull().sum().idxmax()
```

```python
#Exercise 19
#Find the column with the lowest number of missing values.
students.isnull().sum().idxmin()
```

```python
#Exercise 20
#Calculate dataset completeness percentage.
100-(students.isnull().sum().sum()/ students.size)*100
```

```python
#Exercise 23
#Print all student names using iterrows()
for index, row in students.iterrows():
    print(row["name"])
```

```python
#Exercise 24
#Print student names and GPA.
for index, row in students.iterrows():
    print(row["name"], row["gpa"])
```

```python
#Exercise 25
#Print students whose GPA is above 3.5.
for index, row in students.iterrows():
    if row["gpa"] > 3.5:
        print(row["name"])
```

```python
#Exercise 26
#Print all scholarship recipients.
for index, row in students.iterrows():
    if row["scholarship"] > 0:
        print(row["name"])
```

```python
#Exercise 27
#Create GPA categories using iteration.
categories = []

for index, row in students.iterrows():
    if row["gpa"] >= 3.7:
        categories.append("Excellent")
    elif row["gpa"] >= 3.0:
        categories.append("Good")
    else:
        categories.append("At Risk")

students["category"] = categories

students["category"]
```

```python
#Exercise 28
#Print all rows using itertuples() .
for row in students.itertuples():

    print(
    row.student_id,
    row.name
    )
```

```python
#Exercise 29
#Count students by department using iteration only.
counts = {}

for index, row in students.iterrows():

    dept = row["department"]
    
    counts[dept] = counts.get(dept, 0) + 1

print(counts)
```

```python
#Exercise 30
#Generate a student summary report.
for index, row in students.iterrows():

    print(f"Student ID: {row['student_id']}")
    print(f"Name: {row['name']}")
    print(f"Department: {row['department']}")
    print(f"GPA: {row['gpa']}")
    print("-" * 30)
```
