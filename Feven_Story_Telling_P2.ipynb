```python
#Exercise 1: Study Hours vs Exam Score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
"hours": [1, 2, 3, 4, 5, 6, 7],
"score": [45, 55, 65, 72, 80, 88, 95]
})
sns.relplot(
    data=df,
    x="hours",
    y="score",
    kind="scatter"
)
plt.title("Study Hours vs Exam Score")
```

```python
#Exercise 2: Student Performance by Gender
import pandas as pd

df = pd.DataFrame({
"hours": [1,2,3,4,5,6,7,8],
"score": [50,60,68,75,80,85,92,98],
"gender": ["Male","Male","Male","Male",

"Female","Female","Female","Female"]

})
sns.relplot(
    data=df,
    x="hours",
    y="score",
    hue="gender",
    style="gender"
)
plt.title("Exam Scores by Gender")
```

```python
#Exercise 3: Assignment Completion
import pandas as pd

df = pd.DataFrame({
"hours": [1,2,3,4,5,6],
"score": [48,58,67,74,83,90],
"assignments": [2,4,3,6,5,8]
})
sns.relplot(
    data=df,
    x="hours",
    y="score",
    size="assignments"
)
plt.title("Effect of Assignments on Scores")
```

```python
#Exercise 4: Department Comparison
import pandas as pd

df = pd.DataFrame({
"hours": [1,2,3,4,1,2,3,4],
"score": [50,60,70,80,55,65,75,85],
"department": ["CS","CS","CS","CS","IS","IS","IS","IS"]

})
sns.relplot(
    data=df,
    x="hours",
    y="score",
    col="department"
)
plt.title("Department Comparison")
```

```python
#Exercise 5: Salary Prediction Using Regression
import pandas as pd

df = pd.DataFrame({
"experience": [1,2,3,4,5,6,7,8],
"salary": [30000,35000,42000,50000,
58000,65000,72000,80000]

})
sns.lmplot(
    data=df,
    x="experience",
    y="salary" 
)
plt.title("Experience vs Salary")
```
