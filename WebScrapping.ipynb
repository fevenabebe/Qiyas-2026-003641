```python
import requests
from bs4 import BeautifulSoup
```

```python
html_doc = requests.get('https://www.worldometers.info/coronavirus/')
soup = BeautifulSoup(html_doc.content, 'html.parser')
```

```python
table_content = soup.find('table')
```

```python
table_head = table_content.find('thead')
table_body = table_content.find('tbody')
table_row = table_body.find_all('tr')
table_data = table_body.find_all('td')
```

```python
file = open("tests.csv", "w", encoding="utf-8")

headers = [th.get_text(strip=True) for th in table_head.find_all('th')]
file.write(','.join(headers) + '\n')

for row in table_row[8:]:
    row_data = [td.get_text(strip=True) for td in row.find_all('td')]
    file.write(','.join(row_data) + '\n')
    print(row_data)
file.close()
```

```python
file = open("test.csv", "w", encoding="utf-8")

headers = [th.get_text(strip=True) for th in table_head.find_all('th')]
file.write(','.join(headers) + '\n')

for row in table_row[8:]:
    row_data = [td.get_text(strip=True).replace(',', '') for td in row.find_all('td')]
    file.write(','.join(row_data) + '\n')
    print(row_data)
file.close()
```

```python
file = open("tests.csv", "w", encoding="utf-8")

headers = [th.get_text(strip=True) for th in table_head.find_all('th')]
file.write(','.join(headers) + '\n')

for row in table_row[8:]:
    row_data = []

    for td in row.find_all('td'):
        value = td.get_text(strip=True).replace(',', '')

        if value.replace('.', '').isdigit():
            value = f'="{value}"'

        row_data.append(value)

    file.write(','.join(row_data) + '\n')
    print(row_data)

file.close()
```
