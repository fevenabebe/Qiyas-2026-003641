```python
#import library
import matplotlib.pyplot as plt
```

```python
#Exercise 1: Monthly Sales Trend
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

plt.plot(months,sales, marker="s", markersize=8, color="Green", linestyle="--" , label='Sales')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Montly Sales Trend")
plt.legend()
plt.show()
```

```python
#Exercise 2: Student Scores
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]

plt.plot(subjects,scores,color="red", marker="*", markersize=9, linewidth=3, markerfacecolor="yellow",label="Scores")
plt.xlabel("Subject")
plt.ylabel("Scores")
plt.title("Exam score")
plt.legend()
plt.show()
```

```python
#Exercise 3: Company Revenue
years = [2021, 2022, 2023, 2024]
revenue = [500, 700, 850, 1000]

plt.plot(years,revenue, color="blue",linestyle="dotted",linewidth=2 ,marker="d",label="Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.title("Revenue Per Year")
plt.legend()
plt.show()
```

```python
#Exercise 4: Fruit Supply (Vertical Bar Chart)
fruits = ['Apple', 'Orange', 'Banana', 'Avocado']
counts = [50, 70, 40, 60]
colors = ['red', 'orange', 'yellow', 'green']

fig, ax = plt.subplots()
bars=ax.bar(fruits,counts,color=colors)
ax.set_title("Fruit Supply")
ax.set_ylabel("Quantity")
ax.set_xlabel("Fruits")
ax.legend(bars,fruits)
plt.show()
```

```python
#Exercise 5: Stacked Bar Chart (Students by Gender)
import numpy as np

departments = ('CS', 'IS', 'SE', 'IT')

students = {
'Male': np.array([70, 65, 80, 50]),
'Female': np.array([60, 55, 70, 45])
}
fig, ax= plt.subplots()
male= ax.bar(departments, students["Male"], width=0.7, label="Male" )
female=ax.bar(departments, students["Female"], width=0.7, label="Female", bottom=students["Male"])
ax.bar_label(male, label_type='center')
ax.bar_label(female, label_type='center')

ax.set_title('Students by Department and Gender')
ax.legend()

plt.show()
```

```python
import pandas as pd

sales=pd.DataFrame({
    "day":["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday", "Sunday"],
    "cups_solid":[120, 135,110,145,160,210,190]
})

colors = ["grey"] * len(sales)
max_index = sales["cups_solid"].idxmax()
colors[max_index] = "red"

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(sales["day"], sales["cups_solid"], color=colors)

ax.bar_label(bars, padding=3)


ax.set_title("Weekly Sales (Saturday Highlighted as Max)", fontsize=14, pad=15)
ax.set_ylabel("Cups Sold")
ax.set_ylim(0, 250) 

plt.show()
```
