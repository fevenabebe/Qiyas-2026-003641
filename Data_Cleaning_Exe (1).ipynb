```python
import pandas as pd
import numpy as np

university=pd.DataFrame({
    "student_id":[201,202,203,204,205,206,207,208,209,210],
    "name":["Abel", "Sara", "None", "Jhon", "Martha", "None", "David", "Helen", "Tom", "None"],
    "department":["CT","IT","CS", "None","SE","IT", "SE", "None","CS","IT"],
    "gpa":[3.5, 3.8, np.nan, 2.9, 3.7, np.nan, 3.2, 3.4, 3.9, np.nan],
    "scholarship": [5000, np.nan, 3000, 2000, np.nan, 4000, 3500, np.nan, 5000, np.nan]
})
```

```python
Missing_value=university.isnull()
Missing_value
```

```python
#Count missing values
Missing_sum= university.isnull().sum()
Missing_sum
```

```python
#calculate missing-value percentage
percentage = Missing_sum/len(university)*100
percentage
```

```python
#Replace missing names with unknown
university["name"]=university["name"].fillna("Unknown")
university["name"]
```

```python
#Replace missing department with mode
university["department"]=university["department"].fillna(university["department"].mode()[0])
university["department"]
```

```python
#Replace missing gpa values with median
university["gpa"]=university["gpa"].fillna(university["gpa"].median())
university["gpa"]
```

```python
#Replace missing scholarship with zero
university["scholarship"]=university["scholarship"].fillna(0)
university["scholarship"]
```

```python
#Data Analysis
```

```python
#Count students by department
counts = {}

for index, row in university.iterrows():

    dept = row["department"]
    
    counts[dept] = counts.get(dept, 0) + 1
print(counts)
```
