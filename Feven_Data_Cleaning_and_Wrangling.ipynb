# Exercise 11: Outlier Detection

```python
df = pd.DataFrame({
"salary":[3500,4000,4200,3900,4100,45000]
})
```

### Calculate Q1 and Q3

```python
Q1= df["salary"].quantile(0.25)
Q3=df["salary"].quantile(0.75)
print(Q1,"\n")
print(Q3)
```

Calculate IQR

```python
IQR= Q3 - Q1
print(IQR)
```

Identify outliers

```python
lower = Q1 - 1.5 * IQR
upper= Q3 + 1.5 * IQR
outliners = df[(df["salary"] < lower) | (df["salary"] > upper)]
outliners
```

Remove outliers

```python
Clean_outliners_df= df[(df["salary"]>= lower) & (df["salary"]<= upper)]
Clean_outliners_df
```

# Exercise 12: Phone Number
# Standardization

```python
df = pd.DataFrame({
"phone":[
"0911223344",
"+251911223344",
"251911223344",
"0911-223-344"
]
})
```

Remove symbols

```python
df["phone"]= df["phone"].str.replace("-","", regex=False)
df["phone"]=df["phone"].str.replace("+","",regex=False)
df
```

Convert all numbers to +251 format

```python
df["phone"]=df["phone"].apply(
    lambda x: "+251" + x[1:] if x.startswith("0") else "+251" + x[3:]
)
df
```

# Exercise 15: Mixed Data Cleaning
# Challenge

```python
#Dataset
import numpy as np
import pandas as pd

df = pd.DataFrame({
"name":[" Abel ","Sara",None,"JOHN"],
"age":[20,np.nan,25,30],
"salary":[4000,4500,5000,45000],
"join_date":[
"2026-01-01",
"01/02/2026",
"March 10, 2026",
"2026.04.15"
]
})
```

Clean names.


```python
df["name"]=df["name"].fillna("Unknown").str.strip().str.lower().str.title()
df
```

Fill missing values

```python
df["age"]=df["age"].fillna(df["age"].mean())
df
```

Standardize dates


```python
df["join_date"] = pd.to_datetime(df["join_date"],format="mixed")
df
```

Detect outliers

```python
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[
(df["salary"] < lower) |
(df["salary"] > upper)
]

print(df)

print("Outliers")

print(outliers)
```
