```python
def Generate_list(text):
 
    text = text.replace(",", " ").replace(".", " ")
    word_list = text.lower().split()
    sorted_words = sorted(word_list)
    return sorted_words
```

```python
Generate_list(text)
```

```python
def Generate_dictinary(text):
 
    text = text.replace(",", " ").replace(".", " ")
    word_list = text.lower().split()
    sorted_words = sorted(word_list)
    return count_words(sorted_words)

def count_words(word_list):
    dct = {}
    for word in word_list:
        if word in dct:
            dct[word] = dct[word] + 1
        else:
            dct[word] = 1
    return dct
```

```python
Generate_dictinary(text)
```
